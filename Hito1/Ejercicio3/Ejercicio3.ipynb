Diseña un Python que consuma gran volumen de datos y lo muestre visualmente
(gráfico, dashboard…)

La base de datos usada es un archivo csv sobre casos de violencia con armas en estados unidos. https://raw.githubusercontent.com/AdrianPlaza17/SGE_H1/main/US-Gun-Violence.csv

```python
import pandas as pd

df=pd.read_csv(r"https://raw.githubusercontent.com/AdrianPlaza17/SGE_H1/main/US-Gun-Violence.csv")
print(df)
```

```python
import matplotlib.pyplot as plt
 
x = df['state']
y = df['killed']+df['injured']

fig, ax=plt.subplots()

ax.scatter(x,y)

plt.ylabel('Fallecidos + Lesionados')
plt.xlabel('Estados')

plt.xticks(rotation='vertical')
plt.rcParams['figure.figsize'] = [12, 10]
plt.show()


```

Esta grafica tipo scattered muestra el numero de afectados por casos de ataques con armas en diferentes estados de EEUU.
